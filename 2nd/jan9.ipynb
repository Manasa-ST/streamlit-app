# CASE STUDY OF CHOLERA


```python
import geopandas as gpd
import contextily as ctx
```

```python
deaths_df = gpd.read_file('SnowGIS/Cholera_Deaths.shp')
pumps_df = gpd.read_file('SnowGIS/Pumps.shp')
```

```python
deaths_df.head()
```

```python
deaths_df.crs
```

```python
pumps_df
```

Plotting the Outbreak

```python
ax=deaths_df.plot(column='Count',alpha=0.5,edgecolor='k',legend=True) #plot is alpha , egde color is outline , legend is scale
```

```python
ax = deaths_df.plot(column ='Count',figsize=(10,10),alpha=0.5,edgecolor='k',legend=True)
pumps_df.plot(ax=ax,marker='x',markersize=50)
```

```python
ax = deaths_df.plot(column ='Count',figsize=(15,15),alpha=0.5,edgecolor='k',legend=True)
pumps_df.plot(ax=ax,marker='x',markersize=50)
ctx.add_basemap(ax, crs=deaths_df.crs.to_string(), ) #import contextily as ctx. basemap
```

```python
ax = deaths_df.plot(column ='Count',figsize=(15,15),alpha=0.9,edgecolor='blue',legend=True)
pumps_df.plot(ax=ax,marker='x',color='red',markersize=100);

ctx.add_basemap(ax, crs=deaths_df.crs.to_string(),source="SnowGIS/SnowMap.tif") 

```
