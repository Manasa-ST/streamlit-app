```python
import seaborn as sns
import matplotlib.pyplot as plt
data = [ 10,20,30,23,43,23]
sns.histplot(data)
plt.title("Histogram - Seaborn")
plt.show()
```

```python
names = ["A", "B", "C"]
marks = [80, 90, 85]

plt.bar(names, marks)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Bar Chart - Matplotlib")
plt.show()

import pandas as pd

df = pd.DataFrame({
    "Students": ["A", "B", "C"],
    "Marks": [80, 90, 85]
})

sns.barplot(x="Students", y="Marks", data=df)
plt.title("Bar Chart - Seaborn")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [50, 55, 65, 70, 75]

plt.scatter(x, y)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Scatter - Matplotlib")
plt.show()
```

```python
sns.scatterplot(x=x, y=y)
plt.title("Scatter - Seaborn")
plt.show()
```
