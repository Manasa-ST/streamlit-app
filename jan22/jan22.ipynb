```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
clrimage = cv2.imread('eye.jpg')

# Convert to RGB
clrimage_rgb = cv2.cvtColor(clrimage, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 4))

# BGR image (incorrect colors)
plt.subplot(1, 2, 1)
plt.imshow(clrimage)
plt.title('BGR Image (OpenCV)')
plt.axis('off')

# RGB image (correct colors)
plt.subplot(1, 2, 2)
plt.imshow(clrimage_rgb)
plt.title('RGB Image (Matplotlib)')
plt.axis('off')

plt.show()
```

```python
image1 = cv2.imread('eye.jpg')
rgb_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
_, binary_image = cv2.threshold(gray_image1, 127, 255, cv2.THRESH_BINARY)
```

```python
print("----Image Properties----")
print("Color Image Shape(H, W, C):", rgb_image1.shape)
print("Grayscale Image Shape(H, W):", gray_image1.shape)
print("Image data type:", rgb_image1.dtype)
print("Total number of pixels in Color Image:", rgb_image1.size)
print("Pixel value Range",rgb_image1.min(),"to", rgb_image1.max())
print("Average Pixel Intensity:", np.mean(image1))
```

Display images using subplots

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(rgb_image1)
plt.title('Color Image (RGB)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gray_image1, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')
plt.axis('off')

plt.show()
```

CONVERSION OF IMAGES

```python
image2 = cv2.imread('eye.jpg')
rgb_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB) #bgr -> rgb
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY) #bgr -> gray
hsv_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV) #bgr -> hsv
lab_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2Lab) #bgr -> lab
YCrCb_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2YCrCb) #bgr -> ycrcb
gray_bgr_image2 = cv2.cvtColor(gray_image2, cv2.COLOR_GRAY2BGR) #gray -> bgr
add_alpha = cv2.cvtColor(image2, cv2.COLOR_BGR2BGRA) #add alpha channel

plt.figure(figsize=(15, 10))

# Original BGR
plt.subplot(3, 3, 1)
plt.imshow(image2)
plt.title('BGR Image')
plt.axis('off')

# RGB
plt.subplot(3, 3, 2)
plt.imshow(rgb_image2)
plt.title('RGB Image')
plt.axis('off')

# Grayscale
plt.subplot(3, 3, 3)
plt.imshow(gray_image2, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# HSV 
plt.subplot(3, 3, 4)
plt.imshow(cv2.cvtColor(hsv_image2, cv2.COLOR_HSV2RGB))
plt.title('HSV Image')
plt.axis('off')

# LAB 
plt.subplot(3, 3, 5)
plt.imshow(cv2.cvtColor(lab_image2, cv2.COLOR_Lab2RGB))
plt.title('LAB Image')
plt.axis('off')

# YCrCb 
plt.subplot(3, 3, 6)
plt.imshow(cv2.cvtColor(YCrCb_image2, cv2.COLOR_BGR2YCR_CB))
plt.title('YCrCb Image')
plt.axis('off')

# Gray to BGR
plt.subplot(3, 3, 7)
plt.imshow(gray_bgr_image2)
plt.title('Gray  BGR')
plt.axis('off')

# BGRA (alpha channel)
plt.subplot(3, 3, 8)
plt.imshow(add_alpha)
plt.title('BGRA Image')
plt.axis('off')

plt.show()

```

```python
#USe case : Detecting specific color in the image
image5 = cv2.imread('traffic_light_006.jpg')
image6 = cv2.cvtColor(image5, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(image5, cv2.COLOR_BGR2HSV)

lower_red = np.array([0, 120, 70])
upper_red = np.array([10, 255, 255])

mask = cv2.inRange(hsv, lower_red, upper_red)
result = cv2.bitwise_and(image5, image5, mask=mask)
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image6)
plt.title('input image in RGB')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(mask)
plt.title('Input image of red masked')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(result)
plt.title('Result Image')
plt.axis('off')
plt.show()
```

```python
#USe case : Detecting specific color in the image
image5 = cv2.imread('traffic_light_006.jpg')
image6 = cv2.cvtColor(image5, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(image5, cv2.COLOR_BGR2HSV)

#red            yellow         \Green
#0,120,70    20.120.100     32,120,70
#10,255,255  30,255,255     50,255,255
lower_red = np.array([20,120,100])
upper_red = np.array([30,255,255])

mask = cv2.inRange(hsv, lower_red, upper_red)
result = cv2.bitwise_and(image5, image5, mask=mask)
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image6)
plt.title('input image in RGB')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(mask)
plt.title('Input image of red masked')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(result)
plt.title('Result Image')
plt.axis('off')
plt.show()
```
